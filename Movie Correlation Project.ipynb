Mean

```python
#Calculate mean of score (or see as .describe())

avg_score = df['score'].astype('float').mean(axis=0)  #astype for changing type to make sure it will show float(0.000)
print('Avarage of score: ', avg_score)
```

```python
#Replace 'NaN' with mean of score
df['score'].replace(np.nan, avg_score, inplace=True)
```

```python
#Check whether replaced (or check after change all)
df['score'].isnull().value_counts()
```
